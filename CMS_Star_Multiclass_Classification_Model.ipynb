Model Pipeline:
Utilizing Pipeline and GridsearchCV, loop through various classification models(including neural network) as well as hyperparameters while also doing cross validation to select the optimal model.

Conclusion:
Discuss at a high level the results of the optimal model.

# User Defined Function(UDF)

## Column Analyzer

```python
def column_analyzer(df):
    
    # This function loops through columns in a dataframe returning the number of unique values,
    # the number of null records, and the top 5 values in each column based on frequency
        
    #data types
    data_types = df.dtypes
    
    #Unique values
    uniques = df.nunique()
    
    #Percent of nulls
    percent_of_nulls = round(df.isna().sum()/len(df)*100,0)
    
    #combine pandas series
    df_column_analyzer = pd.concat([data_types, uniques, percent_of_nulls], axis=1)
    
    #reset index
    df_column_analyzer = df_column_analyzer.reset_index()
    
    #rename columns
    df_column_analyzer.columns = ['Field', 'Data_Type', 'Unique_Values', 'Percent_of_Nulls']
    
    #intialize top 5 frequent values dataframe
    top_5_df = pd.DataFrame(columns = ['Field','Value_1', 'Percent_of_Total_1',
                                   'Value_2', 'Percent_of_Total_2',
                                   'Value_3', 'Percent_of_Total_3',
                                   'Value_4', 'Percent_of_Total_4',
                                   'Value_5', 'Percent_of_Total_5'
                                   ])
    
    #get list of column names in df
    col_list = list(df.columns)

    for i in col_list:
        #Get a series of values and convert counts to percent of total
        top_5 = round(df[i].value_counts().sort_values(ascending=False)/len(df)*100,0)

        #Pass to dictionary
        top_5.to_dict()

        #select top 5 items in dictionary
        top_5 = {k: top_5[k] for k in top_5.keys()[:5]}

        #intialize list to hold values and percents
        top_5_list = []

        #first append column name
        top_5_list.append(i)

        #append column value and percent
        for j in top_5:
            top_5_list.append(j)
            top_5_list.append(top_5[j])
        
        #check if list is too short and if so append nan
        if len(top_5_list) < 11:
            items_needed = 11 - len(top_5_list)

            for k in range(items_needed):
                top_5_list.append(np.nan)
        else:
            pass


        #insert list into data frame
        top_5_df.loc[len(top_5_df), :] = top_5_list

    #join dataframes on left join to create final output
    final_columnizer = pd.merge(df_column_analyzer, top_5_df, how="left", on=["Field", "Field"])

    #erturn the final table
    return final_columnizer
```

## SQL Extractor

```python
def sql_extract(sql):
    
    #MySQL host name
    hostname="localhost"
    
    #MySQL database name, note that i pass database name in actual SQL, so this is just a placeholder
    dbname="cms_prd"
    
    #read in username from text file
    with open('MySQL_usnm.txt', 'r') as f:
        uname = f.read()
        
    #read in password from text file
    with open('MySQL_pwd.txt', 'r') as f:
        pwd = f.read()
    
    #create SQLAlchemy engine to connect to MySQL Database
    engine = create_engine("mysql+pymysql://{user}:{pw}@{host}/{db}".format(host=hostname,
                                                                            db=dbname,
                                                                            user=uname,
                                                                            pw=pwd))
    
    #connect to MySQL
    dbConnection = engine.connect()
    
    #extract data from MySQL and pass to pandas df
    sql_data = pd.read_sql(sql, dbConnection)
    
    #close conneciton to MySQL
    dbConnection.close()
    
    #return pandas df
    return sql_data
```

## SQL Importer

```python
def sql_import(df,schema,tablename):
    
    #MySQL host name
    hostname = "localhost"
    
    #MySQL database name
    dbname = schema
    
    #read in username from text file
    with open('MySQL_usnm.txt', 'r') as f:
        uname = f.read()
        
    #read in password from text file
    with open('MySQL_pwd.txt', 'r') as f:
        pwd = f.read()
    
    #create SQLAlchemy engine to connect to MySQL Database
    engine = create_engine("mysql+pymysql://{user}:{pw}@{host}/{db}".format(host=hostname,
                                                                            db=dbname,
                                                                            user=uname,
                                                                            pw=pwd))
    
    #replaces spaces in column headers with "_"
    df.columns = df.columns.str.replace(' ', '_')
    
    #Replaces special characters with nothing
    df.columns = df.columns.str.replace('[^\w\s]','')
    
    #connect to MySQL
    dbConnection = engine.connect()
    
    #send pandas dataframe to MySQL table
    df.to_sql(tablename, engine, index=False)
    
    #close conneciton to MySQL
    dbConnection.close()
    
    #return pandas df
    return
```

## One Hot Encoder(single column)

```python
def custom_ohe(df, df_column):
    #gets unique values in order from largest to smallest
    value_list = df[df_column].value_counts().index.tolist()
    
    #gets number of unique values
    value_length = len(value_list)
    
    #intializes ohe value list
    ohe_values = []
    
    #loops through value length appending value to ohe_values
    for i in range(value_length):
        ohe_values.append(i)

    #create dictionary between lists to be used to remap column
    ohe_dict = dict(zip(value_list, ohe_values))
    
    #print dic to screen for visibility
    print(ohe_dict)
    
    #updates column with OHE values
    df.replace({df_column: ohe_dict},inplace=True)
    
    return
```

## Alpha Remover and Mean Replacer

```python
def alpha_remove_mean_replace(df, column):
    #Removes alphas and keeps decimals
    df[column] = df[column].replace('[^\d.]', '', regex=True)

    #replaces blank fields with np.nan
    df[column] = df[column].replace(r'^\s*$', np.nan, regex = True)

    #converts fields to numeric
    df[column] = df[column].astype(float)

    #fill np.nan with averages
    df[column].fillna(value=df[column].mean(), inplace=True)
    
    return
```

## CMS API Extractor

```python
def cms_api_extractor(dataset_id):

    #API link
    url = 'https://data.cms.gov/provider-data/api/1/metastore/schemas/dataset/items/' + dataset_id

    #get API link results
    url_results = requests.get(url, verify = False)

    #extract json from results
    results = json.loads(url_results.content)

    #extract API data download link from json results
    #Note that the API download link is dynamic based on dates and is updated quarterly
    cms_csv = results['distribution'][0]['downloadURL']

    #send API data to pandas df
    cms_df = pd.read_csv(cms_csv, encoding = 'latin1')
    
    return cms_df
```

## Initialize Neural Network

```python
# define Neural Network model - Shallow with three layers, using n_variables from feature selection
def create_model():
    # create model
    model = Sequential()
    model.add(Dense(8, input_dim=n_variables, activation='relu'))
    model.add(Dense(5, activation='softmax'))
    # Compile model
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```
